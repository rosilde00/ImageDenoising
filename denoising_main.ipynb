```python
from keras.layers import Dense, Input
import matplotlib.pyplot as plt
from keras.models import Model
import pandas as pd
import numpy as np

#Build del modello
input_layer = Input(shape=(784,))
encode_layer1 = Dense(600, activation='relu')(input_layer)
encode_layer2 = Dense(100, activation='relu')(encode_layer1)
latent_view   = Dense(20, activation='sigmoid')(encode_layer2)
decode_layer1 = Dense(100, activation='relu')(latent_view)
decode_layer2 = Dense(600, activation='relu')(decode_layer1)
output_layer  = Dense(784)(decode_layer2)

model = Model(input_layer, output_layer)
model.summary()

# Ottenimento dei pesi
model.load_weights('./results')

#Ottenimento delle prime 5 immagini del dataset per la predizione
images = pd.read_csv("./fashion-mnist_train.csv")
images = images.drop('label', axis=1)
images = images[0:5]
clean_images = images[list(images.columns)].values
clean_images = clean_images / 255
clean_images = clean_images.reshape(-1, 784)

#Noising delle immagini
noisy_images = []
for i in range(0, len(clean_images)):
    x = np.random.rand(784)
    noise = np.random.normal(x, 0.2*x)
    noisy_img = clean_images[i] + noise
    noisy_images.append(noisy_img)
noisy_images = np.asarray(noisy_images)

#Predizione sul test set
predictions = model.predict(noisy_images)

#Plot del test set con noise
f, ax = plt.subplots(1,5)
f.set_size_inches(80, 40)
for i in range(5):
    ax[i].imshow(noisy_images[i].reshape(28, 28))
plt.show()

#Plot del denoising delle immagini del test set
f, ax = plt.subplots(1,5)
f.set_size_inches(80, 40)
for i in range(5):
    ax[i].imshow(predictions[i].reshape(28, 28))
plt.show()

#Plot delle immagini del test set prima dell'applicazione del noise
f, ax = plt.subplots(1,5)
f.set_size_inches(80, 40)
for i in range(5):
    ax[i].imshow(clean_images[i].reshape(28, 28))
plt.show()
```
